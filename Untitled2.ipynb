## Helper func

```python
def fix_dollar(x):
  x=str(x)
  x=x.replace(')', '') 
  x=x.replace('(', '') 
  x=x.replace('$', '')
  if 'M' in x:
    x=float(x.replace('M', ''))
    x=x*1000000
    return x
  return x

def fix_eth(x):
  x=str(x)
  x=x.replace('Ξ', '')
  x=float(x)
  return x
```
